```python
# Trying Xpath  and proper scripting type
def checking_on_codingninja_story():
         
    list_stories=(driver.find_elements_by_xpath('//*[@id="react-root"]/section/main/section/div[1]/div[1]/div/div/div/div/ul/li'))[2:]
    list_canvas=driver.find_elements_by_xpath('//*[@id="react-root"]/section/main/section/div[1]/div[1]/div/div/div/div/ul/li//canvas')
    
    count=0
    for story in list_stories:
        
        if story.text=="coding.ninjas":
            if list_canvas[count].get_attribute("height")=="10":
                print("Story has been opened there! check the session")
                story.click()
            else:
                print("Story has been Seen!")
            break
        count+=1
    else:
        print("Story Not there")

checking_on_codingninja_story()
```
